# Statistics

Learning more about **statistics package**

## importing packages

```python
import statistics as st
import random
import matplotlib.pyplot as plt

import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
```

##  Random data generation

```python
data = []
for i in range(100):
    x = random.randint(1,60)
    data.append(x)
```

## Getting data from API

```python
from libs.importData import getSpecificValue 
data = getSpecificValue(["PETR4"])[0]["close"]
```

## Calculation statistics metrics

```python
mean = st.mean(data)
median = st.median(data)
pstdev = st.pstdev(data)
stdev = st.stdev(data)

from tabulate import tabulate
metrics_header = ["mean", "median", "pstdev", "stdev"]
metrics_data = [['%5.2f' % mean, '%6.2f' % median, '%5.2f' % pstdev, '%5.2f' % stdev]]
print(tabulate(metrics_data, headers=metrics_header))
```
